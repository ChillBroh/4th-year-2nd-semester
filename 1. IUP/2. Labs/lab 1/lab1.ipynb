```python
import cv2
import matplotlib.pyplot as plt

print("1. fetch image from laptop" )
image = cv2.imread("/Users/isharakankanamge/Downloads/SLIIT/3. IUP/2. Labs/lab 1/images.jpeg")
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.title("My Image")
plt.axis('on')
plt.show()
```

```python
print("2. Access pixel values and change" )
px = image2[100,10]
print("[100,10] value is", px)
image2[100:110, 50:60] = [255, 255, 255]
px2 = image2[100,10]
print("[100,10] new value is", px2)
plt.imshow(image2)
plt.title("My Image with removed white part")
plt.axis('on')
plt.show()
```

```python
print("3. Access image1 and image2 properties")
print("Shape (Height, Width, Channels):", image.shape)
print("Size (Total Pixels):", image.size)
print("Data Type:", image.dtype)

print("Shape (Height, Width, Channels):", image2.shape)
print("Size (Total Pixels):", image2.size)
print("Data Type:", image2.dtype)
```

```python
print("4.Split and merge channels")
b, g, r = cv2.split(image)
merged = cv2.merge((b,g,r))

plt.subplot(1, 3, 1)
plt.title('Blue Channel')
plt.imshow(b, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Green Channel')
plt.imshow(g, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Red Channel')
plt.imshow(r, cmap='gray')

```

```python
print("5.Resize and Rotate image")
# Resize image
resized = cv2.resize(image, (300, 300))  # Width x Height
print("new size",image.shape)
# Rotate 90 degrees clockwise
rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
```

```python
# Draw a blue rectangle
cv2.rectangle(image2, (50, 50), (200, 200), (255, 0, 0), 2)
# Draw a red circle
cv2.circle(image2, (300, 300), 50, (0, 0, 255),6)
# Draw a green line
cv2.line(image2, (0, 0), (300, 300), (255, 255, 0), 5)
plt.imshow(image2)
```

```python
cv2.putText(image, 'Hello OpenCV', (10, 500), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)
plt.imshow(image)
```

```python

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Binary (thresholding)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Negative
negative = cv2.bitwise_not(image)

# 1. Grayscale
plt.subplot(1, 3, 1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('on')

# 2. Binary
plt.subplot(1, 3, 2)
plt.imshow(binary, cmap='gray')
plt.title("Binary")
plt.axis('on')

# 3. Negative
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(negative, cv2.COLOR_BGR2RGB))
plt.title("Negative")
plt.axis('on')

```
