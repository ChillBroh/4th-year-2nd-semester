```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the image
img = cv2.imread('building.jpg',0)
```

```python
#Apply Sobel operator / sobel filter
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 3)#Sobel(img,flag,order)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 3)
```

```python
#scaling
sobelxAbs = cv2.convertScaleAbs(sobelx)
sobelyAbs = cv2.convertScaleAbs(sobely)
```

```python
#Diplay the results
plt.imshow(img, 'gray')
plt.show()

plt.imshow(sobelx, 'gray')
plt.show()

plt.imshow(sobely, 'gray')
plt.show()

plt.imshow(sobelxAbs, 'gray')
plt.show()

plt.imshow(sobelyAbs, 'gray')
plt.show()
```

```python
import cv2

# Load the image in grayscale
image = cv2.imread('building.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian Blur to reduce noise
blurred_image = cv2.GaussianBlur(image, (3, 3), 0)

# Apply Laplacian operator
laplacian = cv2.Laplacian(blurred_image, cv2.CV_64F)

# Convert the result to uint8
laplacian = np.uint8(np.absolute(laplacian))

# Display the result
cv2.imshow('Laplacian', laplacian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
2. Laplacian Operator
Laplacian operator is aslo a derivative operator which is used to find edges in an image. 
The difference between Laplacian and other operators is that others are first order derivative masks 
but Laplacian is a second order derivative mask. In this mask we have two further classifications one is 
Positive Laplacian Operator and other is Negative Laplacian Operator.

Another difference between Laplacian and other operatord is that unlike others Laplacian doesn't take 
edges in any particular direction but it takes edges as Inward edges and Outward edges
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

#Load the image
img = cv2.imread('building.jpg',0)


#Apply Laplacian operator / Laplacian filter
laplacian = cv2.Laplacian(img, cv2.CV_64F)#laplacian without scaling Laplacian(original img, convertion flag)
#consider responses on positive and negative
laplacianAbs = cv2.convertScaleAbs(laplacian)#laplacian with scaling


#Diplay the results
plt.imshow(img, 'gray')
plt.show()

plt.imshow(laplacian, 'gray')
plt.show()


plt.imshow(laplacianAbs, 'gray')
plt.show()
```
