### 2. Power-law transformation to improve the contrast of a dark image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread("D:/SLIIT Y4 S1/IUP/Labs/Lab 5/My/gamma.jpg",0)

gamma_b = 0.4
img_power_b = np.power(img1, gamma_b)

gamma_d = 3
img_power_d = np.power(img1, gamma_d)


plt.imshow(img1, cmap= 'gray')
plt.show()

# bright image
plt.imshow(img_power_b, cmap= 'gray')
plt.show()

# dark image
plt.imshow(img_power_d, cmap= 'gray')
plt.show()

cv2.waitKey(0)
```

### Log transformation to improve the dynamic range of an image 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img3 = cv2.imread("D:/SLIIT Y4 S1/IUP/Labs/Lab 5/My/log.jpg",0)

# Handling deviding by zero issue and also recieving negative values.
np.seterr(divide = "ignore")
np.seterr(invalid = "ignore")

# Get the constant value through calculation.
c = 255/np.log(1 + np.max(img3))

# Apply equation
# Transform the image ( into uint8)
log_img = c * (np.log(1 + img3)) 

# Convert the image into intensity levels from uint8 type
log_img = np.array(log_img, dtype = np.uint8)

print(np.max(img3))


plt.imshow(img3, cmap = 'gray')
plt.show()

plt.imshow(log_img, cmap = 'gray')
plt.show()

print(c)

cv2.waitKey(0)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale mode
img3 = cv2.imread("D:/SLIIT Y4 S1/IUP/Labs/Lab 5/My/log.jpg", 0)

# Check if image is loaded correctly
if img3 is None:
    raise FileNotFoundError("Error: Image not found. Check file path.")

# Set error handling for warnings
np.seterr(divide="ignore", invalid="ignore")

# Prevent log(0) issue and check max value
max_value = np.max(img3)
if max_value == 0:
    max_value = 1  # Prevent division by zero

# Compute the scaling constant
c = 255 / (np.log(1 + max_value + 1e-10))

# Ensure img3 is non-negative to avoid log(negative)
img3 = np.maximum(img3, 0)

# Apply log transformation
log_img = c * np.log(1 + img3)

# Convert to uint8
log_img = np.array(log_img, dtype=np.uint8)

# Display original and transformed images
plt.imshow(img3, cmap="gray")
plt.title("Original Image")
plt.show()

plt.imshow(log_img, cmap="gray")
plt.title("Log Transformed Image")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("D:/SLIIT Y4 S1/IUP/Labs/Lab 5/My/log.jpg",0)

# Check if the image is loaded successfully
if image is None:
    print("Error: Image not found or unable to load.")
else:
    # Avoid log(0) by shifting the image values
    image = np.float32(image)  # Convert to float to handle potential non-integer values
    c = 255 / np.log(1 + np.max(image))  # Calculate scaling constant
    log_image = c * np.log(1 + image)  # Apply log transform

    # Clip values to be in the valid range of 0-255
    log_image = np.clip(log_image, 0, 255)

    # Convert to uint8
    log_image = np.array(log_image, dtype=np.uint8)

    # Display both images
    plt.imshow(image, cmap='gray')
    plt.show()
    plt.imshow(log_image, cmap='gray')
    plt.show()
   
    cv2.waitKey(0)
```
