### 2) Histograam Calculation / plotting in Numpy

```python
img = cv2.imread('D:/SLIIT Y4 S1/IUP/Labs/Lab 3/My/sunflower.jpg',1)

b,g,r = cv2.split(img)

# Alternative way to find histogram of an image
hist2,bins = np.histogram(b.ravel(),256,[0,256])
hist3,bins = np.histogram(g.ravel(),256,[0,256])
hist4,bins = np.histogram(r.ravel(),256,[0,256])

plt.plot(hist2)
plt.plot(hist3)
plt.plot(hist4)

plt.show()
```

## 3) Create a hitogram for a specific area in the image

```python
mask = np.zeros(img.shape[0:2], np.uint8)  # creating a full black mask
mask[50:300, 250:400] = 255

plt.imshow(mask) # Normally the mask is generated in the B G R format. So that we have to convert it to a black and white one
```

```python
mask = np.zeros(img.shape[0:2], np.uint8)  # creating a full black mask
mask[50:300, 250:400] = 255

plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)) # convert the mask into black and white
```

```python
hist5 = cv2.calcHist([mask], [0], None, [256], [0,256])
plt.plot(hist5)
plt.show()
```

## Extrct the image of the selected area into the mask

```python
maskedImage = cv2.bitwise_and(img, img, mask = mask)
plt.imshow(cv2.cvtColor(maskedImage, cv2.COLOR_BGR2RGB))
```

### Create the histogram for the selected area of the image

```python
color = ('b','g','r')

for i, col in enumerate(color):
    hist6 = cv2.calcHist([img], [i], mask, [256], [0,256]) # The parameter 256 is the number of bins. You can chang it as you want
    plt.plot(hist6, color = col)
    plt.show()
```
